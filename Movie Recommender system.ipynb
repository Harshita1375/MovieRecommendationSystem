```python
df_new=df.dropna()
```

```python
df.columns
```

```python
df_new.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10840, stop_words='english')
```

```python
vector = cv.fit_transform(df_new['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
df_new[df_new['original_title']=='Jurassic World'].index[0]
```

```python
distance = sorted(list(enumerate(similarity[0])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(df_new.iloc[i[0]].original_title)
```

```python
def recommend(movies):
    index= df_new[df_new['original_title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(df_new.iloc[i[0]].original_title)
    
```

```python
recommend("Batman")
```
